```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi


def generate_Digital_Signal(bsig):
  sym_len=50

  rand_n=bsig
  
  no_bit=np.size(rand_n)

  #Generate Signal

  sig=np.zeros(sym_len*no_bit)

  id_n=[]
  for i,j in enumerate(rand_n):
    if j=='1':
      id_n.append(i)

  for i in id_n:
    temp1= int(i*sym_len)
    sig[temp1:temp1+sym_len]=1

  return sig




inp= input("Enter a bit string: ")
rand_n=list(inp)

sig=generate_Digital_Signal(rand_n)

space_separated = " ".join(rand_n)

plt.figure(figsize=(20,5))
plt.plot(sig,"blue")
plt.title('Digital signal')
plt.xlabel(space_separated,fontsize=15)
```

```python

plt.figure(figsize=(20,11))

n= np.arange(-10,11,1)
l=np.size(n)
impulse=np.zeros(l)
for i,j in enumerate(n):
  # print("i= " ,i ,"j= ",j)
  if j==0 :
    impulse[i]=1
    break
plt.subplot(2,1,1)
plt.stem(n,impulse)
plt.title('Impulse Signal')
plt.xlabel('Sample'); plt.ylabel('Amplitude')


stp=np.zeros(l)
for i,j in enumerate(n):
  # print("i= " ,i ,"j= ",j)
  if j>=0 :
    stp[i]=1

plt.subplot(2,1,2)
plt.stem(n,stp)
plt.title('Step Signal')
plt.xlabel('Sample'); plt.ylabel('Amplitude')

plt.tight_layout
```
