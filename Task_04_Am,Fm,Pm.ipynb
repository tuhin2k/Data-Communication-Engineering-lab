```python
"""

Reg: 2017331101
Name: Asif Idris Tuhin

"""

"""

Amplitude Modulated signal, s(t)= Ac*cos(2π*fc*t)*[1 + Mo*cos(2π*fm*t)] Where fm= The frequency of messeage signal, fc= The frequency of carrier signal, Mo= Modulation index(Am/Ac), Ac= The amplitude of the carrier signal

"""


import matplotlib.pyplot as plt
import numpy as np
from math import pi



Fs = 10000; #Sampling Frequency
# Time axis
t = np.arange(0,2,1/Fs)

# Generate Carrier wave
Fc = 50;   #Carrier Signal Frequency
Ac=1 # Carrier signal amplitude

c = Ac*np.cos(2*pi*Fc*t) # Carrier wave


# Generate message signal
Fm = 5; #message signal frequency
Am = 2 #messeage signal amplitude
m = Am*np.cos(2*pi*Fm*t) # message signal


# Amplitude modulated signal
Mo=Am/Ac
s = c * (1 + Mo * np.cos(2*pi*Fm*t)) 

# plot

plt.figure(figsize=(20,15))
plt.suptitle("Amplitude Modulation (AM)",fontsize=18)

plt.subplot(4,1,1)
plt.plot(t,m,"blue")
plt.xlabel("Analog Message Signal",fontsize=13)

plt.subplot(4,1,2)
plt.plot(t,c,"red")
plt.xlabel("Carrier signal",fontsize=13)

plt.subplot(4,1,3)
plt.plot(t,s,"green")
plt.xlabel("Modulated signal",fontsize=13)

plt.subplot(4,1,4)
plt.plot(t,s)
plt.plot(t,m)
plt.xlabel("Message Signal and Modulated signal in one plot for better understanding",fontsize=13)
plt.legend(['Amplitude Modulated Signal','Message Signal'])

plt.subplots_adjust(hspace=1)
```

```python
"""

Frequency Modulated signal, s(t) = Ac*cos(2π*fc*t+Mo*sin(2π*fm*t)) Where fm= The frequency of messeage signal, fc= The frequency of carrier signal, Mo= Modulation index, Ac= The amplitude of the carrier signal

"""



import matplotlib.pyplot as plt
import numpy as np
from math import pi

Fs = 10000; #Sampling Frequency
Fc = 100;   #Carrier Signal Frequency
Fm = 15; #message signal frequency
Am = 2 #messeage signal amplitude
Ac=1 # Carrier signal amplitude
#Time Axis
t = np.arange(0,0.5,1/Fs)
Mo = 4; # Modulation index

m = Am*np.cos(2*pi*Fm*t) #message signal
c= Ac*np.cos(2*pi*Fc*t) #carrier signal

s = Ac*np.cos(2*pi*Fc*t + Mo*np.sin(2*pi*Fm*t)) #modulated signal


#plot

plt.figure(figsize=(20,15))
plt.suptitle("Frequency Modulation (FM)",fontsize=18)

plt.subplot(4,1,1)
plt.plot(t,m,"blue")
plt.xlabel("Analog Message Signal",fontsize=13)

plt.subplot(4,1,2)
plt.plot(t,c,"red")
plt.xlabel("Carrier signal",fontsize=13)

plt.subplot(4,1,3)
plt.plot(t,s,"green")
# plt.plot(t,m)
plt.xlabel("Modulated signal",fontsize=13)

plt.subplot(4,1,4)
plt.plot(t,s)
plt.plot(t,m)
plt.legend(['Frequency Modulated Signal','Message Signal'])

plt.xlabel("Message Signal and Modulated signal in one plot for better understanding",fontsize=13)


plt.subplots_adjust(hspace=1)
```

```python
"""

Phase Modulated Signal, s(t) = Ac*cos(2π*fc*t+Mo*cos(2π*fm*t)) Where fm= The frequency of messeage signal, fc= The frequency of carrier signal, Mo= Modulation index, Ac= The amplitude of the carrier signal

"""



import matplotlib.pyplot as plt
import numpy as np
from math import pi

Fs = 10000; #Sampling Frequency
Fc = 100;   #Carrier Signal Frequency
Fm = 15; #message signal frequency
Ac= 1 # Carrier signal amplitude
Am = 1 #messeage signal amplitude

#Time Axis
t = np.arange(0,0.5,1/Fs)
Mo = 4; # Modulation index

m = Am*np.cos(2*pi*Fm*t) #message signal
c= Ac*np.cos(2*pi*Fc*t) #carrier signal


s = Ac*np.cos(2*pi*Fc*t + Mo*np.cos(2*pi*Fm*t)) # Modulated Signal


#plot

plt.figure(figsize=(20,15))
plt.suptitle("Phase Modulation (PM)",fontsize=18)

plt.subplot(4,1,1)
plt.plot(t,m,"blue")
plt.xlabel("Analog Message Signal",fontsize=13)

plt.subplot(4,1,2)
plt.plot(t,c,"red")
plt.xlabel("Carrier signal",fontsize=13)

plt.subplot(4,1,3)
plt.plot(t,s,"green")
# plt.plot(t,m)
plt.xlabel("Modulated signal",fontsize=13)

plt.subplot(4,1,4)
plt.plot(t,s)
plt.plot(t,m)
plt.legend(['Phase Modulated Signal','Message Signal'])
plt.xlabel("Message Signal and Modulated signal in one plot for better understanding",fontsize=13)


plt.subplots_adjust(hspace=1)
```
