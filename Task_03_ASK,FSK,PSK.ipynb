```python
"""

Reg No: 2017331101
Name: Asif Idris Tuhin

"""

import matplotlib.pyplot as plt
import numpy as np
from math import pi

def gen_DigitalSignal(number_of_bits,sample_per_bit):
  #number_of_bits: Number of Bits to be generated
  #sample_per_bit: Number of Samples per Bit
  
  #generate Random numbers between 0 and 1
  rand_numbers=np.random.rand(number_of_bits)


  for i,j in enumerate(rand_numbers):
    if j >= 0.5:
      rand_numbers[i]=1
    else:
      rand_numbers[i]=0

  signal=np.zeros(number_of_bits*sample_per_bit)

  #generate bits
  indx=np.where(rand_numbers == 1)

  for i in indx[0]:
    temp=int(i*sample_per_bit)
    signal[temp:temp+sample_per_bit]=1
  
  return signal,rand_numbers



Fs=1000 #sampling frequency for both digital and carrier signal. 
fc=40 # carrier frequency
a=1 # carrier amplitude
T=2 # total simulation time 

t= np.arange(0,T,1/Fs) 

x=a*np.sin(2*pi*fc*t) #carrier signal

#Generating Digital signal
Td=0.1 # Bit duration 
sample_per_bit=int(Td*Fs) # 1 second e sample nicchi Fs . so 1bit er jonno time lage Td so Td time e sample nite hobe Fs*Td . orthat etogula sample/point nicchi to represent 1 bit
number_of_bits=   int(np.floor(np.size(t) / sample_per_bit))

rand_number=[]
digital_signal,rand_number=gen_DigitalSignal(number_of_bits,sample_per_bit)



temp=[] #for priting bitstring in title
for i in rand_number:
  temp.append(int(i))

bitstring = ' '.join([str(elem) for elem in temp]) #list to string



# print("Random Bit string: ",rand_number)

#Generate ASK

ask=x*digital_signal


#plot
plt.figure(figsize=(20,15))
plt.suptitle("Amplitutde Shift Keying",fontsize=18)
plt.subplot(4,1,1)
plt.plot(t,digital_signal,"blue")
plt.title("Random Bit string: " + bitstring,fontsize=13)
plt.xlabel("Digital Message Signal",fontsize=11)

plt.subplot(4,1,2)
plt.plot(t,x,"red")
plt.xlabel("Carrier signal",fontsize=11)


plt.subplot(4,1,3)
plt.plot(t,ask,"green")
plt.xlabel("Modulated signal",fontsize=11)

plt.subplot(4,1,4)
plt.plot(t,ask)
plt.plot(t,digital_signal)
plt.xlabel("Digital Message Signal and Modulated signal in one plot for better understanding",fontsize=11)
plt.legend(['Modulated Signal','Digital Message Signal'])



plt.subplots_adjust(hspace=1)

```

```python



#generate FSK
f=fc+2*fc*digital_signal/2
fsk=np.sin(2*pi*f*t)


# print("Random Bit string: ",rand_number)
#plot
plt.figure(figsize=(20,15))
plt.suptitle("Frequency Shift Keying",fontsize=18)
plt.subplot(4,1,1)
plt.plot(t,digital_signal,"blue")
plt.title("Random Bit string: " + bitstring,fontsize=13)
plt.xlabel("Digital Message Signal",fontsize=11)

plt.subplot(4,1,2)
plt.plot(t,x,"red")
plt.xlabel("Carrier signal",fontsize=11)


plt.subplot(4,1,3)
plt.plot(t,fsk,"green")
plt.xlabel("Modulated signal",fontsize=11)


plt.subplot(4,1,4)
plt.plot(t,fsk)
plt.plot(t,digital_signal)
plt.xlabel("Digital Message Signal and Modulated signal in one plot for better understanding",fontsize=11)
plt.legend(['Modulated Signal','Digital Message Signal'])



plt.subplots_adjust(hspace=1)
```

```python
#generate PSK
d= digital_signal*180

psk=np.sin(2*pi*fc*t + d )

# print("Random Bit string: ",rand_number)
#plot
plt.figure(figsize=(20,15))
plt.suptitle("Phase Shift Keying",fontsize=18)
plt.subplot(4,1,1)
plt.plot(t,digital_signal,"blue")
plt.title("Random Bit string: " + bitstring,fontsize=13)
plt.xlabel("Message Digital Signal",fontsize=11)

plt.subplot(4,1,2)
plt.plot(t,x,"red")
plt.xlabel("Carrier signal",fontsize=11)


plt.subplot(4,1,3)
plt.plot(t,psk,"green")
plt.xlabel("Modulated signal",fontsize=11)


plt.subplot(4,1,4)
plt.plot(t,psk)
plt.plot(t,digital_signal)
plt.xlabel("Digital Message Signal and Modulated signal in one plot for better understanding",fontsize=11)
plt.legend(['Modulated Signal','Digital Message Signal'])





plt.subplots_adjust(hspace=1)
```
